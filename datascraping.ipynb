```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.icc-cricket.com/"
response = requests.get (url)
if response.status_code == 200:
    soup = BeautifulSoup(response.content,"html.parser")
    title = soup.title.string
    print(f"Title of the page : {title}")
    links = soup.find_all("a")
    for link in links:
        print(link.get("href"))
else:
    print(f"failed to receivethe webpage . status code : {response.status_code}")
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')
title = soup.find('title')  # Find the <title> tag
print(title.text)  # Extract and print text from the <title> tag
headings = soup.find_all('h2')  # Find all <h2> tags
for heading in headings:
    print(heading.text)
specific_div = soup.find_all('div', class_='specific-class')
for div in specific_div:
    print(div.text)
links = soup.find_all('a')  # Find all <a> tags (links)
for link in links:
    print(link.get('href'))  # Extract the href attribute (URL)
images = soup.find_all('img')  # Find all <img> tags (images)
for image in images:
    print(image.get('src'))  # Extract the src attribute (image URL)
import csv

links = soup.find_all('a')
with open('links.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Link'])  # Write header
    for link in links:
        writer.writerow([link.get('href')])
```
